```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly
import json
import seaborn as sns
from matplotlib.pyplot import figure
import datetime
import plotly.express as px
from sklearn.cluster import KMeans
import plotly.graph_objs as go
import numpy as np
```

```python
def to_time(x):
    if (x > 4) and (x <= 8):
        return 'Early Morning'
    elif (x > 8) and (x <= 12 ):
        return 'Morning'
    elif (x > 12) and (x <= 16):
        return'Noon'
    elif (x > 16) and (x <= 20) :
        return 'Eve'
    elif (x > 20) and (x <= 24):
        return'Night'
    elif (x >=0)and(x <= 4):
        return'Late Night'
```

```python
df=pd.read_csv("USvideos.csv")
```

```python
with open('US_category_id_2.json', 'r') as f:
    data = json.load(f)
```

```python
cat_id={}
for i in data['items']:
    cat_id[i['id']]=i['snippet']['title']
```

```python
category_id_df=pd.DataFrame(data=cat_id.items(),columns=["category_id","category"])
```

```python
category_id_df['category_id']=category_id_df['category_id'].astype("int")
```

```python
df=df[df['likes']!=0]
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
vid_id=list(set(df['video_id'].values))
max_view=0
df_=pd.DataFrame()
for vid in vid_id:
    views=df[df['video_id']==vid]['views'].values
    max_view=max(views) if len(views)>1 else views[0]
    row=df[(df['video_id']==vid)&(df['views']==max_view)]
    df_=pd.concat([df_,row])
```

```python
df=df_
```

```python
df.reset_index(drop=True,inplace=True)
df.drop_duplicates(subset=['video_id'],inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
df=df.merge(category_id_df,on='category_id',how='inner')
```

```python
df
```

```python
df['publish_time']=df['publish_time'].apply(lambda x:pd.to_datetime(x))

df=df.sort_values(by='publish_time',ascending=False)

df.reset_index(drop=True,inplace=True)

df['publish_time_day']=df['publish_time'].dt.hour

df['publish_time_day']=df['publish_time_day'].apply(lambda x:to_time(x))



weekday_map={0:"monday",1:"tuesday",2:"wednesday",3:"thursday",4:"friday",5:"saturday",6:"sunday"}
df['publishedAt_day']=df['publish_time'].apply(lambda x:weekday_map[datetime.datetime.weekday(x)])
```

```python
df['publishedAt_day_time']=df['publishedAt_day'].astype(str)+" "+df["publish_time_day"].astype(str)
```

```python
df['likes_dislikes_ratio']=df['likes']/(df['dislikes']+1)
```

```python
df.drop_duplicates(subset=['video_id'],inplace=True)
```

```python
df['response_type']=df['likes_dislikes_ratio'].apply(lambda x:"good response"if x>=2 else "bad response")
```

```python
df['publish_year']=df['publish_time'].dt.year
df['publish_hour']=df['publish_time'].dt.hour
```

```python
fig=px.histogram(df, x="category",color='response_type')
fig.update_layout(xaxis={'categoryorder':'total descending'})

fig.show()
```

- Most of the trending videos are from **Entertainment,** **Music,**  **Howto & Style** and **News & Politics**
- We will analyze only **top 3** trending video categories

#### Side Note

- **Bad response** is when No.of dislikes is **almost equal** to No.of likes or when No.of dislikes is **more than** No.of likes

- A video can trend either because of **positive reasons** or for **negative reasons**

- From the above graph, we see that **News & Politics** has **more bad responses** compared to other categories.

------------------------------------------------------------------------------------------------------------------------------

```python
fig=px.histogram(df, x="publish_hour",color='response_type')
fig.update_layout(xaxis={'categoryorder':'total descending'})

fig.show()
```

- Most of the trending videos were published between **2 pm (14:00:00) to 6 pm (18:00:00)**
- Irrespective of the time, if the **video content is not good** then the **response** of the viewers will be **bad**. We can see the bad response of the viewers in all published hour.
